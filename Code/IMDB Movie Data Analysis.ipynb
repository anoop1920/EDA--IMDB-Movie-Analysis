### 18. Classify Movies Based on Ratings 

```python
data.columns
```

```python
def rating(rating):
    if rating >=7.5:
        return "Excellent"
    elif rating >=6.0:
        return "Good"
    else:
        return "Average"
```

```python
# Create new column of rating category in our dataset
data['rating_category'] = data['Rating'].apply(rating)
```

### 20 . Find Unique Values From Genre

```python
data['Genre']
```

```python
# Make a list for all the items
list1 = []
for item in data['Genre']:
    list1.append(item.split(','))
```

```python
list1
```

```python
# convert list1 into 1-D list to find unique values
one_D = []
for item in list1:
    for item in item:
        one_D.append(item)
```

```python
one_D
```

```python
Unique_list = []
for item in one_D:
    if item not in Unique_list:
        Unique_list.append(item)
```
